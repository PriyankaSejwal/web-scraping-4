Q2. Scrape the details team India’s international fixtures from bcci.tv.You need to find following details:
A) Match title (I.e. 1
st ODI)
B) Series
C) Place
D) Date
E) Time

```python
driver=webdriver.Chrome(r'C:\Users\Priyanka\Documents\chromedriver.exe')   #connnecting to the web driver
```

```python
driver.get('https://www.bcci.tv/')  #opening the url given
time.sleep(2)
driver.maximize_window()
```

```python
# import Action chains  
from selenium.webdriver.common.action_chains import ActionChains 

#xpath for the option international
international=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]")
#xpath to choose fixtures
fixture=driver.find_element_by_xpath("//div[@class='drop-down__options']//li[1]//a")

#using action chains to click on the fixture tab hidden inside the internatinal box
ActionChains(driver).move_to_element(international).click(fixture).perform()
```

```python
fixtures={}
fixtures['Match_title']=[]
fixtures['Series']=[]
fixtures['Place']=[]
fixtures['Time']=[]
fixtures['full_date']=[]

#extracting Match title from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/strong"):
        fixtures['Match_title'].append(i.text)
except NoSuchElementException:
    fixtures['Match_title'].append("--")
        

#extracting name of series from xpath
try:
    for i in driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
        fixtures['Series'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Series'].append("--")

    
#extracting place from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span"):
        fixtures['Place'].append(i.text)
except NoSuchElementExceptions:
    fixtures['Place'].append("--")
    

#extracting time from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[2]"):
        fixtures['Time'].append(i.text)
except NoSuchElementException:
    fixtures['Time'].append("--")
    
    
#extracting date and month from xpath
date=[]
try:
    for i in driver.find_elements_by_xpath("//div[@class='fixture__full-date']/span"):
        date.append(i.text)
except NoSuchElementException:
    date.append("--")
month=[]
try:
    for i in driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[1]"):
        month.append(i.text)
except NoSuchElementException:
    month.append("--")

#combining date and month
for i in range(len(month)):
    fixtures['full_date'].append(date[i]+month[i])
```

```python
fixtures_df=pd.DataFrame.from_dict(fixtures)
fixtures_df
```

```python
fixtures_df.to_csv('fixture.csv')
```

```python
driver.close()
```
